## Cleaning and Recreating
The data has brackets in GEO and Violations, we need to clean it some more by splitting and creating a new chart.

```python
#I have to remove the brackets in the Violations column, so I split it
cleanedviolations = crime['Violations'].str.split('[',n=1,expand=True)
cleanedviolations
```

```python
#Now violations is cleaned with no brackets, saved in crime table
crime['Violations']=cleanedviolations[0]
crime
```

```python
#I have to remove the brackets in the GEO column, so I split it
cleanedgeo = crime['GEO'].str.split('[',n=1,expand=True)
cleanedgeo
```

## Updated Table
Now table CRIME is now cleaned successfully 

```python
#Adding the tables together
crime['GEO']=cleanedgeo[0]
crime
```

# Location

Now we go into Locations with the total incidents of crime, we have to group them by location and sum them to get a exact total. I have organized it by ascending to see which is the highest.

```python
#Grouping by GEO to get a understanding of each location, I then sumed the value of each incident. 
#I also changed the column name to Total Incident.
crimegeo = crime.groupby ('GEO').sum()['VALUE'].to_frame()
crimegeo.columns = ['Total Incidents']
crimegeo = crimegeo.sort_values ('Total Incidents', ascending=False)
crimegeo
```

## Location Outcome

We have determined that my hypothesis is true, Ontario has the highest incidents of Crime for Provinces.

# Crime

Now we go into crime types with the total violations of crime, we have to group them by violations and sum them to get a exact total. I have organized it by ascending to see which is the highest.

```python
#I have grouped it by Violations and found the sum of the values. 
#I have also adjusted the column name to Total Violations
crimeviolations = crime.groupby ('Violations').sum()['VALUE'].to_frame()
crimeviolations.columns = ['Total Violations']
crimeviolations = crimeviolations.sort_values ('Total Violations', ascending=False)
crimeviolations
```

## Crime Outcome

We have determined that my hypothesis is false, Assault has the highest incidents of Crime.

# Exploratory data analysis (Statistics)

Here we have a understanding of the outcome of each crime, we can see how many of them were cleared, charged youth or adult.

```python
#I grouped by statistics and found the sum of the value.
#I have adjusted the column name to be Total Statistics.
crimestatistics = crime.groupby ('Statistics').sum()['VALUE'].to_frame()
crimestatistics.columns = ['Total Statistics']
crimestatistics = crimestatistics.sort_values ('Total Statistics', ascending=False)
crimestatistics
```

# Exploratory data analysis (Year)

Here we have a understanding of the outcome of each year, I have grouped them by year and calculated the total crime per year.

```python
#I have grouped by REF_DATE and found the sum of the value.
#I have also changed the column name to Total Crime.
crimeyear = crime.groupby ('REF_DATE').sum()['VALUE'].to_frame()
crimeyear.columns = ['Total Crime']
crimeyear
```
